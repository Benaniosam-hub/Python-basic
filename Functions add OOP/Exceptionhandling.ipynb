16.⁠ ⁠What is the difference between try, except, else, and finally?

```python
try = handling the error which is gonna happen, try passes the error to execpt 
execpt = checkes the error, 
common exceptions:
zeroDivisionError
valueError
noFileFoundError
typeError

else = if try runs without error then it pass to else 

finally = is whether the code is error or not this comment allow to continue the code {variable_name.close()}
```

```python
def withdraw_money():
    print("ATM starts")
    account_balance = 5000  
    
    try:
        
        amount = int(input("Enter the amount: "))
        if amount > account_balance:
            raise ValueError("out of bank balance!")
            
        result = account_balance - amount

    except ValueError as error_msg:
       
        print(f"Access denied: {error_msg}")
        
    else:
        
        print("success")
        print(f"Account balance: Rs. {result}")
        
    finally:
        
        print("Take your atm card")
    

withdraw_money()
```
