Create a base class called person with a constructor that takes a name as a parameter. create a derived class called student that inherits from person and constructor that takes a parameter called grade. Write a method in student to display the name and grade.

```python
class person:
    def __init__(self,name):
        self.name=name
        print(f"Your name is: {name}")
class student(person):
    def __init__(self,name,grade):
        self.name=name
        self.grade=grade    
        print(f"Your name is: {name}, Your grade is: {grade}")
stud1=student("Benanio","A+")

    
```
