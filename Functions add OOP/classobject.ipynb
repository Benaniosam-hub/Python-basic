10. What is the difference between a class and an object?

```python
# a class is a blueprint or a constructor to create an object
# a object is a instance of that class,created from the blueprint(class)
class laptop():
    def __init__(self,col):
        self.color=col
        self.ram="which ram"

    def display(self):
        print("color:",self.color)
        print("Ram:",self.ram)
        
hp=laptop("red") # -> passing the valur through parameter
lenovo=laptop("black")

hp.ram="5gb"
lenovo.ram="8gb"

hp.display()
lenovo.display()
```
