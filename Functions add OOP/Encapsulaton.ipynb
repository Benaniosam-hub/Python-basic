Encapsulation

```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.__salary=salary
       
    def get_salary(self):
        return self.__salary
        
    def set_salary(self,new_salary):
        if new_salary > 0:
            new_salary=self.__salary
        else:
            print("Wrong salary")


emp=Employee("bena",5000)

print(emp.get_salary())


emp.set_salary=6000
print (emp.set_salary)
        
        
```
