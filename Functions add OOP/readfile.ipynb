17.⁠ ⁠How do you read a file in Python?


to read a file, use "r" inside the open() statement.
3 methods to read a file
file.read() -> reads the entire file
file.readline() -> reads only the first line of the content
file.readlines() -> reads all the content into list 
file.read(5) -> specific content by index 

```python
with open ("/Users/sam/my_repositor/Python-basic/.ipynb_checkpoints/geeks.txt", "r") as file:
    print(file.read())
    print(file.readline())
    print(file.readlines())
```
