15.⁠ ⁠How do you handle exceptions in Python?


In python handling exception is using "try" and "except" blocks catches the error and manages the error safetly without crashing the application.
the core structure : try,except,else,finally

```python
try:
    a = int(input("enter the number: "))

    if a == 0:
        print("don't divide by zero")

    else:
        b = 10 / a
        v=int(b)
        print(f"Output: {v}")

except ValueError:
    print("Please enter numbers only")
    
```
