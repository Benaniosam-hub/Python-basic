```python
list_a=[1,2,3,4,5,3,2]
result=list(set([x for x in list_a if list_a.count(x) > 1]))
print ("Duplicates:",result)
```

```python
a=[1,2,3,7,7,7,7,7,8,8]
d=list(set([x for x in a if a.count(x) > 1]))
print ("Duplicates: ",d)

```
