23.⁠ ⁠Write a program to count the occurrence of each character in a string.
" AI is strong and AI is smart "

```python
sentence=(" AI is strong and AI is smart ").lower()
unique=set(sentence)

for i in unique:
    if i == " ":
        continue
    count= sentence.count(i)
    print(f"{i} : {count} times")
```
