21.⁠ ⁠Write a program to find duplicate elements in a list.


```python
a=[1,2,3,4,3,2,5]
remove_duplicate=set()
duplicate=[]

for value in a:
    if value in remove_duplicate:
        duplicate.append(value)
        
    else:
        remove_duplicate.add(value)

print("Duplicate: ",duplicate)    
```
