2.What is the difference between a list and a tuple?


```python
# a list is ordered and used for storing multiple values in one variable and can be modified later which is mutable.
# a tuple is also ordered collection of multiple values in one variable but immutable which cant make any change after modified.

my_list=[20,"sam",0.01] # -> syntax in square bracket
print(my_list[0]) # -> 20 as output -> index and slicing method, can changable

you_tuple=(0,"sam",0.002) # syntax in parentheses
you_tuple[2] # -> 0.002 as output -> we can use index and slicing method only for calling the values, but cant be modified
```

3.What is the difference between == and is?


```python
# "==" is used for comparison and checking whether the values are same.
tuple1 =(0,1,2)
tuple2 =(0,1,2)
print(tuple1 == tuple2) # -> True cause both are equal 

# "is" uesd to check both are same address memory then it gives a true comment 
# eg:
list1=[1,1,2,4]
list2=[2,1,4,5]
print(list1 is list2) # -> False cause both are stored in different memory
# but,
list3=[1,2,3,4]
new_list=list3
print(list3 is new_list) # -> True cause both have the same memory address

```

 6.⁠ ⁠What is list comprehension? Give an example. (new to me)


```python
# list comprehension is a concise way to create a new list by adding expression to an existing iterables like list,tuple,range,string

first_list=[x for x in range(13) if x%2 == 0]
print(first_list)
second_list=[y for y in range(14) if y%2 != 0]
print(second_list)

a=[2, 3, 4, 5]
l1=[x**2 for x in a]
print(l1)


```
